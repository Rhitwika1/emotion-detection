```python
# Import necessary libraries
import cv2  # OpenCV for image and video processing
from keras.models import model_from_json  # To load model architecture from a .json file
import numpy as np  # For numerical operations like array handling
from IPython.display import display, clear_output  # Useful if running in Jupyter Notebook (not essential here)

# Load the pre-trained model architecture
json_file = open("emotiondetector.json", "r")  # Open the model JSON file
model_json = json_file.read()  # Read the contents of the file
json_file.close()  # Close the file
model = model_from_json(model_json)  # Convert JSON content into a Keras model object
model.load_weights("emotiondetector.h5")  # Load the trained weights into the model

# Load the Haar cascade file for face detection
haar_file = cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'  # Path to Haar Cascade XML file
face_cascade = cv2.CascadeClassifier(haar_file)  # Load the cascade classifier from file

# Define dictionary for mapping prediction index to emotion labels
labels = {0: 'angry', 1: 'disgust', 2: 'fear', 3: 'happy', 4: 'neutral', 5: 'sad', 6: 'surprise'}

# Preprocess input image before feeding it into the model
def extract_features(image):
    feature = np.array(image)  # Convert image to numpy array
    feature = feature.reshape(1, 48, 48, 1)  # Reshape to fit model input (batch, height, width, channel)
    return feature / 255.0  # Normalize pixel values (0-1)

# Start webcam capture (device index 0 is the default webcam)
webcam = cv2.VideoCapture(0)

# Loop continuously to process video frames
while True:
    ret, im = webcam.read()  # Read a frame from the webcam
    gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)  # Convert the image to grayscale for processing
    faces = face_cascade.detectMultiScale(im, 1.3, 5)  # Detect faces in the image

    # Show a warning if more than one face is detected
    if len(faces) > 1:
        cv2.putText(im, "Warning: Multiple users detected!", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
        print("Warning: Only one user allowed for accurate emotion detection.")
    
    try:
        # Loop through all detected faces
        for (p, q, r, s) in faces:
            image = gray[q:q+s, p:p+r]  # Crop the face region from the grayscale image
            cv2.rectangle(im, (p, q), (p+r, q+s), (255, 0, 0), 2)  # Draw a rectangle around the face
            image = cv2.resize(image, (48, 48))  # Resize face to 48x48 pixels for the model
            img = extract_features(image)  # Preprocess the image
            pred = model.predict(img)  # Predict the emotion using the model
            prediction_label = labels[pred.argmax()]  # Get the label of the highest confidence prediction

            # Display the predicted emotion above the rectangle
            cv2.putText(im, '% s' % prediction_label, (p-10, q-10), cv2.FONT_HERSHEY_COMPLEX_SMALL, 2, (0, 0, 255))

        cv2.imshow("Output", im)  # Show the final image with bounding boxes and labels

        # Break the loop if 'q' is pressed
        if cv2.waitKey(27) & 0xFF == ord('q'):
            break
            
    except cv2.error:
        # If there's a CV error (like empty face crop), skip to next frame
        pass

# Release the webcam and close any OpenCV windows
webcam.release()
cv2.destroyAllWindows()


#flow of the code
#Flow of the Entire Program:
#Model Loading:
#
#Loads a pre-trained emotion recognition CNN model (json + h5).
#
#Face Detection Setup:
#
#Loads Haar Cascade for detecting faces from live webcam feed.
#
#Live Webcam Feed:
#
#Starts capturing video using your webcam in real time.
#
#Face Detection in Frames:
#
#For every video frame, detects faces using the Haar Cascade.
#
#Multi-user Check:
#
#If more than one face is found, a warning is shown on-screen and in console.
#
#Emotion Prediction:
#
#Crops the face.
#
#Converts it to grayscale, resizes to 48x48, reshapes and normalizes it.
#
#Passes the face image to the model.
#
#Displays the predicted emotion label on screen near the face.
#
#Exit Option:
#
#Press q to stop the webcam and exit the window.
```
